```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("dataset/Unemployment_Rate_upto_11_2020.csv")
print(df.head())
print(df.info())
print(df.isnull().sum())
print(df.columns)
plt.figure(figsize=(12,6))
sns.histplot(df.iloc[:, -1], kde=True)
plt.title("Unemployment Distribution")
plt.show()
plt.figure(figsize=(12,6))
sns.lineplot(x=df.iloc[:,0], y=df.iloc[:,-1])
plt.xticks(rotation=90)
plt.title("Unemployment Trend")
plt.show()
plt.figure(figsize=(12,6))
sns.boxplot(x=df.iloc[:,-1])
plt.title("Unemployment Rate Boxplot")
plt.show()
plt.figure(figsize=(12,6))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap="Blues")
plt.title("Correlation Heatmap")
plt.show()
print("Unemployment analysis completed successfully.")
```
